```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow import keras
from tensorflow.keras.layers import Dense,Flatten,Conv2D, MaxPooling2D
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train=x_train/255
x_test=x_test/255
x_train = np.expand_dims(x_train, axis = 3)
x_test = np.expand_dims(x_test,axis = 3)
```

```python
model = keras.Sequential([
    Conv2D(32,(3,3),padding = 'same',activation = keras.activations.relu,input_shape = (28,28,1)),
    MaxPooling2D((2,2),strides = 2),
    Conv2D(64,(3,3),padding = 'same',activation = keras.activations.relu),
    MaxPooling2D((2,2),strides = 2),
    Flatten(),
    Dense(128,activation = keras.activations.relu),
    Dense(10,keras.activations.softmax)])
```

```python
y_train_cat = keras.utils.to_categorical(y_train,10)
y_test_cat = keras.utils.to_categorical(y_test,10)
```

```python
print(model.summary())
```

```python
model.compile(optimizer = 'adam',loss = 'categorical_crossentropy',metrics=['accuracy'])
```
